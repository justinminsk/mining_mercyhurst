## Data

The data contains most important 1-5 gram words based on [tdif score](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html) with no more than 10,000 features. The data also contains hour of day and day of week.


### This data needs to be created with clean_data.py

```python
df = pd.read_parquet("processed_tweets.parquet")
df = df.drop(['time_since_col', 'day_col'], axis=1)
df.head()
```

## Train Test Split

Here we just get x and y data and then randomly split the data to train and test data.

```python
x_data = df.drop(["interaction_target"], axis=1)
y_data = df.interaction_target.values

print(x_data.shape)
print(y_data.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10)

print("x Train:", x_train.shape)
print("y Trian", y_train.shape)
```

## Decision Tree Model
Here is a basic model to find feature importance the depth is 10 and inculded is a:
- list of features from most important to least of features
- a image of the tree
    - this needs to have [graphviz](https://www.graphviz.org/download/) installed 
    - use command line in commments to create `dot -Tpng tree10.dot -o tree10.png`

```python
dtr = DecisionTreeRegressor(criterion='mse', max_depth=10, max_features=None,
           max_leaf_nodes=None, min_impurity_decrease=0.0,
           min_impurity_split=None, min_samples_leaf=2,
           min_samples_split=10, min_weight_fraction_leaf=0.0,
           presort=False, random_state=1, splitter='best')
```

```python
dtr.fit(x_train, y_train)
fi_dict = dict(zip(x_train.columns, dtr.feature_importances_))
fi_dict_1 = { k:v for k, v in fi_dict.items() if v }
fi_dict_sorted = sorted(fi_dict_1.items(), key=lambda kv: kv[1], reverse=True)
fi_dict_sorted
```

```python
dtr.score(x_test, y_test)
```

```python
export_graphviz(dtr, out_file='tree10.dot', 
                feature_names = x_data.columns,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
# call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png'])
# This does not work run : dot -Tpng tree10.dot -o tree10.png

# Display in jupyter notebook
Image(filename = 'tree10.png')
```

## Random Forest Model

This model is a collection of Decision Trees (thus forest). Has simlar features:
- list of fetures from most important to least of features
- a image of the tree
    - this needs to have [graphviz](https://www.graphviz.org/download/) installed

```python
rfr = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=13,
           max_features='auto', max_leaf_nodes=None,
           min_impurity_decrease=0.0, min_impurity_split=None,
           min_samples_leaf=1, min_samples_split=2,
           min_weight_fraction_leaf=0.0, n_estimators=10, n_jobs=1,
           oob_score=False, random_state=1, verbose=0, warm_start=False)
```

```python
rfr.fit(x_train, y_train)
fi_dict = dict(zip(x_train.columns, rfr.feature_importances_))
fi_dict_1 = { k:v for k, v in fi_dict.items() if v }
fi_dict_sorted = sorted(fi_dict_1.items(), key=lambda kv: kv[1], reverse=True)
fi_dict_sorted
```

```python
rfr.score(x_test, y_test)
```

```python
estimator = rfr.estimators_[5]

export_graphviz(estimator, out_file='rf.dot', 
                feature_names = x_data.columns,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
# call(['dot', '-Tpng', 'rf.dot', '-o', 'rf.png'])
# This does not work run : dot -Tpng rf.dot -o rf.png

# Display in jupyter notebook
Image(filename = 'rf.png')
```

## Word2Vec
Here is some code and examples of what word2vec is and how it could be useful.

```python
from gensim.models import Word2Vec
import re
from nltk.corpus import stopwords
# Preprocessing

stopwords = set(stopwords.words('english'))

w2v_df = pd.read_parquet("mm.pq")

pat1 = r'@[A-Za-z0-9_]+'
pat2 = r'https?://[^ ]+'
combined_pat = r'|'.join((pat1,pat2))
www_pat = r'www.[^ ]+'
negations_dic = {"isn't":"is not", "aren't":"are not", "wasn't":"was not", "weren't":"were not",
                "haven't":"have not","hasn't":"has not","hadn't":"had not","won't":"will not",
                "wouldn't":"would not", "don't":"do not", "doesn't":"does not","didn't":"did not",
                "can't":"can not","couldn't":"could not","shouldn't":"should not","mightn't":"might not",
                "mustn't":"must not"}
neg_pattern = re.compile(r'\b(' + '|'.join(negations_dic.keys()) + r')\b')

def pre_processing(row):
    first_process = re.sub(combined_pat, '', row)
    second_process = re.sub(www_pat, '', first_process)
    third_process = second_process.lower()
    fourth_process = neg_pattern.sub(lambda x: negations_dic[x.group()], third_process)
    result = re.sub(r'[^A-Za-z ]','',fourth_process)
    return result.strip()
```

```python
w2v_df.body = w2v_df.body.apply(pre_processing)
```

```python
# create list of words in our tweets
corpus = []
for tweet in w2v_df.body:
    word_list = tweet.split(" ")
    word_list = [w for w in word_list if not w in stopwords]
    corpus.append(word_list)
```

```python
# Create and run model
word2vec = Word2Vec(corpus, min_count=2, size=100, window=10) 
word2vec.train(corpus, total_examples=len(corpus), epochs=10)
```

```python
# http://methodmatters.blogspot.com/2017/11/using-word2vec-to-analyze-word.html
# import the t-SNE library and matplotlib for plotting
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# define the function to compute the dimensionality reduction
# and then produce the biplot
def tsne_plot(model):
    "Creates a TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(8, 8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.savefig("w2v.png")
    plt.show()
    
# call the function on our dataset
tsne_plot(word2vec)
```
