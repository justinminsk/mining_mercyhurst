```python
import logging
import operator
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.externals import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import export_graphviz
from IPython.display import Image
from subprocess import call
```

## Data

The data contains most important 1-5 gram words based on [tdif score](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html) with no more than 10,000 features. The data also contains hour of day and day of week.


### This data needs to be created with clean_data.py

```python
df = pd.read_parquet("processed_tweets.parquet")
df = df.drop(['time_since_col', 'day_col'], axis=1)
df.head()
```

## Train Test Split

Here we just get x and y data and then randomly split the data to train and test data.

```python
x_data = df.drop(["interaction_target"], axis=1)
y_data = df.interaction_target.values

print(x_data.shape)
print(y_data.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10)

print("x Train:", x_train.shape)
print("y Trian", y_train.shape)
```

## Decision Tree Model
Here is a basic model to find feature importance the depth is 10 and inculded is a:
- list of features from most important to least of features
- a image of the tree
    - this needs to have [graphviz](https://www.graphviz.org/download/) installed 
    - use command line in commments to create `dot -Tpng tree10.dot -o tree10.png`

```python
dtr = DecisionTreeRegressor(criterion='mse', max_depth=10, max_features=None,
           max_leaf_nodes=None, min_impurity_decrease=0.0,
           min_impurity_split=None, min_samples_leaf=1,
           min_samples_split=2, min_weight_fraction_leaf=0.0,
           presort=False, random_state=1, splitter='best')
```

```python
dtr.fit(x_train, y_train)
fi_dict = dict(zip(x_train.columns, dtr.feature_importances_))
fi_dict_1 = { k:v for k, v in fi_dict.items() if v }
fi_dict_sorted = sorted(fi_dict_1.items(), key=lambda kv: kv[1], reverse=True)
fi_dict_sorted
```

```python
dtr.score(x_test, y_test)
```

```python
export_graphviz(dtr, out_file='tree10.dot', 
                feature_names = x_data.columns,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
# call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png'])
# This does not work run : dot -Tpng tree10.dot -o tree10.png

# Display in jupyter notebook
Image(filename = 'tree10.png')
```

## Random Forest Model

This model is a collection of Decision Trees (thus forest). Has simlar features:
- list of fetures from most important to least of features
- a image of the tree
    - this needs to have [graphviz](https://www.graphviz.org/download/) installed 
    - use command line in commments to create `dot -Tpng rf.dot -o rf.png`


```python
rfr = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=13,
           max_features='auto', max_leaf_nodes=None,
           min_impurity_decrease=0.0, min_impurity_split=None,
           min_samples_leaf=1, min_samples_split=2,
           min_weight_fraction_leaf=0.0, n_estimators=10, n_jobs=1,
           oob_score=False, random_state=1, verbose=0, warm_start=False)
```

```python
rfr.fit(x_train, y_train)
fi_dict = dict(zip(x_train.columns, rfr.feature_importances_))
fi_dict_1 = { k:v for k, v in fi_dict.items() if v }
fi_dict_sorted = sorted(fi_dict_1.items(), key=lambda kv: kv[1], reverse=True)
fi_dict_sorted
```

```python
rfr.score(x_test, y_test)
```

```python
estimator = rfr.estimators_[5]

export_graphviz(estimator, out_file='rf.dot', 
                feature_names = x_data.columns,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
# call(['dot', '-Tpng', 'rf.dot', '-o', 'rf.png'])
# This does not work run : dot -Tpng rf.dot -o rf.png

# Display in jupyter notebook
Image(filename = 'rf.png')
```
